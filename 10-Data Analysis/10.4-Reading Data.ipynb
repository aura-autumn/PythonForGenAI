A pickle file is a way to serialize and save Python objects — like a frozen snapshot of variables, models, or data — so you can load them later exactly as they were.

File extension: .pkl or .pickle

Purpose: Save and load Python objects in binary format

Library used: pickle (built-in Python module)

Use Case	Example
Save a trained ML model	model.pkl for reuse without retraining
Cache processed data	cleaned_data.pkl to avoid recomputation
Store session or object state	Keep intermediate Python results

You can pickle most Python objects, including:

✅ Lists, dicts, sets
✅ Pandas DataFrames
✅ Trained ML models (like from scikit-learn)
✅ Numpy arrays
✅ Custom Python classes (as long as they’re defined when loading)

```python
df_excel.to_pickle('df_excel')
```

```python
pd.read_pickle('df_excel')
```
